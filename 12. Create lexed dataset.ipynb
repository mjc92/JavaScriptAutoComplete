```python
def custom_lex(text):
    # lexes string, stops when it starts making TypeErrors
    # input: f.read()
    id_list = []
    tok_list = []

    out_list = []
    lexer = Lexer()
    lexer.input(text)
    while True:
        try:
            token = lexer.token()
            if not token:
                break # break if end of token
            if token.type=='ID':
                id_list.append(token.value)
                out_list.append(token.value)
            elif token.type=='STRING':
                out_list.append('STRING')
            elif token.type=='NUMBER':
                out_list.append(token.value)
            else:
                tok_list.append(token.value)
                out_list.append(token.value)
        except TypeError:
            break
        except AttributeError:
            break
    return out_list,id_list,tok_list
```

```python
id_counter = Counter()
tok_counter = Counter()
for i,file in enumerate(file_list):
    with open(os.path.join(file_dir,file).strip()) as f:
        text = f.read()
    out_list, id_list, tok_list=custom_lex(text)
    id_counter += Counter(id_list)
    tok_counter += Counter(tok_list)
    with open(os.path.join(save_dir,'file_%d.txt'%(i+1+startfrom)),'w') as f:
        f.write(' '.join(out_list))
    if (i+1)%100==0:
        print("\n%d files processed so far\n" %(i+1+startfrom))
#     if (i+1)%1000==0:
#         with open('vocabs/id_counter.pckl','wb') as f:
#             pickle.dump(id_counter,f)
#         with open('vocabs/tok_counter.pckl','wb') as f:
#             pickle.dump(tok_counter,f)
#         print("\n%d files collected so far\n" %(i+1+startfrom))
```
