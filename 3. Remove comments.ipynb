```python
import os
import re
from collections import Counter
import slimit
from slimit.parser import Parser
from slimit.visitors.nodevisitor import ASTVisitor
from slimit.lexer import Lexer
from functions import lex, remove_comments
```

```python
# _actual is the result of removing all missing and undecodable files
file_dir = '/home/irteam/users/data/150kJavaScript/train'
lexer = Lexer()

with open(os.path.join(file_dir,'file_list_data_original.txt')) as f:
    file_list=f.read().split('\n')
```

```python
# remove comments
for i,file in enumerate(file_list):
    file = file.strip()
    with open(os.path.join(file_dir,'data_original',file)) as f:
        text = f.read()
    out_text = remove_comments(text)
    with open(os.path.join(file_dir,'data_original',file),'w') as f:
        f.write(out_text)
    if i%1000==0:
        print('%d files done so far' %i)
```

```python
# remove empty lines
for i,file in enumerate(file_list):
    file = file.strip()
    with open(os.path.join(file_dir,'data_original',file)) as f:
        text = f.read()
    lines = text.split('\n')
    out_file = []
    for line in lines:
        line = line.strip()
        if line!='':
            out_file.append(line)
    with open(os.path.join(file_dir,'data_original',file),'w') as f:
        f.write('\n'.join(out_file))
    if i%1000==0:
        print('%d files done so far' %i)
```
