```python
out_idx = 1+out_start_point # fileX
line_count=0 # nth line in a single file
prev_tokens = 0
tokens = []
for i,file in enumerate(file_list[:3000]):
    file = file.strip()
    with open(file_dir+file) as f:
        text = f.read()
    tokens.extend(lex(text,type_value))
        
        
        
#     while len(tokens)>=seq_length:
#         if line_count>=batch_size:
#             line_count=0
#             out_idx+=1
#         else:
#             with open(file_dir+'/batch_%d_seq_%d_%s'
#                       %(batch_size,seq_length,type_value)
#                       +'/file_%d.txt'%out_idx,'a') as f:
#                 f.write(' '.join(tokens[:seq_length])+'\n')
#             tokens = tokens[seq_length:]
#             line_count+=1
#     # when the remaining length is less
#     tmp = tokens + ['<EOS>']
#     for _ in range(seq_length-len(tmp)):
#         tmp = tmp + ['<PAD>']
#     with open(file_dir+'/batch_%d_seq_%d_%s'
#               %(batch_size,seq_length,type_value)
#               +'/file_%d.txt'%out_idx,'a') as f:
#         f.write(' '.join(tmp))
#     line_count+=1
    if (i+in_start_point)%10==0:
        print("====================\n%d files so far!\n====================\n"%(i+in_start_point))
```

```python
token_dict = {}
token_dict['ID']=['ID']
for token in tokens:
    if token.type!='ID':
        if token.type not in token_dict:
            token_dict[token.type]=token.value
```
